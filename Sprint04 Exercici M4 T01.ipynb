```python
#Importació de dades
import pandas as pd

nomfitxer='D:/Users/X/Downloads/preu lloguer per trimestre i barri.csv'
#llegim el fitxer
fitxer=pd.read_csv(nomfitxer,sep=',')
fitxer[:3]
fitxer.head(3)
#fitxer.tail(10)
#fitxer.describe()
```

```python
#Preparació i seleccio de dades de dades en funció els objectius d'estudi

#seleccionem 'Lloguer mitja mensual i el trimestre 1 de l'any i creem un nou dataframe
f=pd.DataFrame(fitxer[(fitxer['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)') & (fitxer['Trimestre']==1)])

#descartem els valors no definits de 'Preu'
f=f[f['Preu']!='--']
#f

#creem una nova columna amb preu com a nombre real
f['Preu_nombre']=f['Preu'].astype(float)


```

```python
#Poden fer un grafic de dispersió per tenir una primera idera dels valors

import matplotlib.pyplot as plt

x_barri=f['Nom_Barri']
#x_barri[:3]
y_preu=f['Preu_nombre']
#y_preu[:3]
fig=plt.figure()
plt.figure(figsize=(15, 6))
plt.scatter(x_barri,y_preu)
plt.title('Any 2021 Trimestre 1: Preu mig per barri')
plt.xlabel('Barris')
plt.ylabel('Preu mig mensual')
plt.xticks(rotation = 45)

#guardem la figura
ruta='D:/Users/X/Downloads/'
plt.savefig(ruta + 'fig1.jpg')

#visualitxem la figura
plt.show()
```

```python
#Volem veure graficament si hi han diferències de preu considerables entre barris
#Ordenar les dades per la variable que volem observar ens ajudara en
#aquesta comparativa


#ordenem per preu
f=f.sort_values('Preu_nombre',ascending=True)
f.tail(100)

x_barri=f['Nom_Barri']
#x_barri[:3]
y_preu=f['Preu_nombre']
#y_preu[:3]
fig2=plt.figure()
plt.figure(figsize=(15, 6))
plt.bar(x_barri,y_preu)
plt.title('Any 2021 Trimestre 1: Preu mig per barri')
plt.xlabel('Barris')
plt.ylabel('Preu mig mensual')
plt.xticks(rotation = 45)

#guardem la figura
ruta='D:/Users/X/Downloads/'
plt.savefig(ruta + 'fig2.jpg')

#visualitxem la figura
plt.show()
```

```python
# Sembla que els barris més cars tenen un comportament destacat sobre
# la resta de barris
# Per constatar-ho, refresentem aquestes dades mitjançant un boxplot
fig3=plt.figure()
plt.figure(figsize=(15, 6))
plt.boxplot(y_preu)
plt.title('Any 2021 Trimestre 1: Preu mig per barri')
plt.xlabel('preus lloguer mensual')
#plt.ylabel('Preu mig mensual')
#plt.xticks(rotation = 45)

#guardem la figura
ruta='D:/Users/X/Downloads/'
plt.savefig(ruta + 'fig3.jpg')

#visualitxem la figura
plt.show()

# Efectivament apareix que els valors mes alts es categoritzen com outliers, que vol dir que estan 1.5 vegades per sobre
# del valor interquartil que conté el 50% dels valors de la mostra
```

```python
#Visualitzció d'una variale numerica: Preu
#seleccionem 'Lloguer mitja mper superficiel i el trimestre 1 de l'any i creem un nou dataframe
df_preu_metre=pd.DataFrame(fitxer[(fitxer['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)') & (fitxer['Trimestre']==1)])
df_preu_metre
#descartem els valors no definits de 'Preu'
df_preu_metre=df_preu_metre[df_preu_metre['Preu']!='--']
#creem una nova columna amb preu com a nombre real
df_preu_metre['Preu_num']=df_preu_metre['Preu'].astype(float)
df_preu_metre.head(3)
```

```python
#Visualitzacio d'una variable numèrica
#Per veure la distribucio de preus decideixo utilitzar un l'histograma
import matplotlib.pyplot as plt
x_preu=df_preu_metre['Preu_num']
#x_barri[:3]
#y_preu=f['Preu_real']
#y_preu[:3]
fig4=plt.figure()
plt.figure(figsize=(15, 6))
plt.hist(x_preu,bins=15)
plt.title('Any 2021 Trimestre 1: Histograma de preu mig per metre')
plt.xlabel('Preu mig per metre')
plt.ylabel('Num  barris')

#guardem la figura
ruta='D:/Users/X/Downloads/'
plt.savefig(ruta + 'fig4.jpg')

#visualitxem la figura
plt.show()
#la distribució s'assembla a una campana de Gauss, lleugerament esbiaxada a la dreta
#possiblement condicionada pels outliers de valors màxims
```

```python
# Volen veure l'evolucio del preu mig mensual a cada districte per trimestre al llarg  del'any . 
# El preu mig a cada trimestre i  acada districte es calculara con la mitja del rpeu mig del lloguer
# dels barris de cada districte

import seaborn as sns

#Preparació de dades
#seleccionem 'Lloguer mitja mensual i el trimestre 1 de l'any, ordenem per preu i creem un nou dataframe
df_districte_preu=pd.DataFrame(fitxer[(fitxer['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)')])
#descartem els valors no definits de 'Preu'
df_districte_preu=df_districte_preu[df_districte_preu['Preu']!='--']
#creem una nova columna amb preu com a nombre real
df_districte_preu['Preu_nombre']=df_districte_preu['Preu'].astype(float)
# Seleccionem les colu mnes que ens interessen per la visualització dessitjada
llista_districte=pd.DataFrame(df_districte_preu[['Trimestre','Nom_Districte','Preu_nombre']])
#agrupen les files de cada districte i trimestre i calculem el preu mig del lloguer als barris del districte
df=pd.DataFrame(llista_districte.groupby(['Nom_Districte','Trimestre']).mean('Preu_nombre'))
df=df.reset_index()

fig5=plt.figure()
plt.figure(figsize=(15, 6))
lp=sns.lineplot(data=df,x='Trimestre',y='Preu_nombre',hue='Nom_Districte')
sns.move_legend(lp,loc="center right")
lp.legend (bbox_to_anchor = (1.2, 1))
```
