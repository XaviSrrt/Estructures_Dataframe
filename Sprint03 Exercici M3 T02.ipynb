El fitxer conte dades de vols de 2008 amb els temps planificats i reals, i amb detall dels retards i a qui són atribuibles
Descripció de les columnes  
-Year: Any  
-Month: Mes  
-DayofMonth: Día del mes  
-DayOfWeek: Día de la setmana; 1=dilluns  
-DepTime: Hora local de sortida; format hhmm  
-CRSDepTime: Hora local de sortida programada; format hhmm  
-ArrTime: Hora local d'arribada, format hhmm  
-CRSArrTime: Hora local d'arribada prevista; format hhmm  
-UniqueCarrier: Codi d'operador únic  
-FlightNum: Número de vol  
-TailNum: Número d'avió  
-ActualElapsedTime: Temps real transcorregut (en minuts)  
-CRSElapsedTime: temps estimat transcorregut (en minuts)  
-AirTime: Temps de vol (en minuts)  
-ArrDelay: Diferencia entre hora d'arribada prevista i la real, en minuts  
           Les arribades anticipades tenen valor negatiu, en minuts  
-DepDelay: Diferència en minuts entre l'hora de sortida prevista i la real  
           Les sortides anticipades mostren números negatius, en minuts  
-Origin: Codi IATA de l'aeroport origen  
-Dest: Codi IATA de l'aeroprt destí  
-Distance: Distància en milles entre sortida i arribada  
-TaxiIn: Baixada de rodes i arribada a la porta de l'aeroport destí, en minuts  
-TaxiOut: Temps transcorregut entre sortida de l'aeroport origen i aixecament de rodes, en minuts  
-Cancelled: Vol cancel·lat?  
-CancellationCode: Motiu de cancel·lació (A = operador, B = temps, C = sistema nacional d'espai aeri, D = seguretat)  
-Diverted: Avió desviat? 1 = sí, 0 = no  
-CarrierDelay: Temps de retard de l'operador, en minuts  
-WeatherDelay: Temps de retard per causes atmosfèriques  
-NASDelay: Temps de retard per causes del sistema nacional del'espai aeri  
-SecurityDelay: Temps de retard per temes de seguretat, en minuts  
-LateAircraftDelay: Temps de retard per causa atribuible a l'avió

```python
#ja que el volum de files amb valors no asignats no es significatiu respecte al volum global, les esborrem
fitxer_selec=fitxer_selec.dropna()
#comptem valors no asignats després d'esborrar
fitxer_selec.isna().sum()
fitxer_selec.shape
```

```python
#Resum estadistic de la mostra

#fitxer_selec.describe()
fitxer_selec[['ArrDelay','AirTime']].describe()
```

```python
#Nova columna VelocitatMitjana(mph)=distancia/temps de vol en hores
fitxer_selec['VelocitatMitjana']=fitxer_selec['Distance']/(fitxer_selec['AirTime']/60)
fitxer_selec
```

```python
#Nova columna tard=veritat si 'ArrTime'>'CRSArrTime'
fitxer_selec['Tard']=fitxer_selec['ArrTime']>fitxer_selec['CRSArrTime']
fitxer_selec
```

```python
fitxer_selec.apply(lambda x: x['Tard'] == True, axis=1).count()
```

```python
print(fitxer_selec.loc[(fitxer_selec['UniqueCarrier']=='AQ') & (fitxer_selec['Tard']==True)])
```

```python
#comptem retards per operador
df_retards_per_operador=pd.DataFrame(fitxer_selec.groupby('UniqueCarrier')['Tard'].sum())
df_retards_per_operador
#comptem vols per operador
df_vols_per_operador=pd.DataFrame(fitxer_selec.groupby('UniqueCarrier')['Tard'].count())
df_vols_per_operador
#fem un inner join entre els dos dataframe per disposar del npombre de vols i de retads per operador
#DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=None, indicator=False, validate=None)[source]
df_vols_retards_operador=df_vols_per_operador.merge(df_retards_per_operador,how='inner',on='UniqueCarrier')
#df_vols_retards_operador.info
#type(df_vols_retards_operador)
df_vols_retards_operador.columns=['Vols','Retards']
#calculem el percentatges de retards
df_vols_retards_operador['Tasa']=df_vols_retards_operador['Retards']/df_vols_retards_operador['Vols']*100
df_vols_retards_operador=df_vols_retards_operador.sort_values(by='Tasa',ascending=False)
#Presentem els 10 operados amb tasa mes alta de retards
df_vols_retards_operador.head(10)
#Presentem els 10 operados amb tasa mes baixa de retards
df_vols_retards_operador.tail(10)
```

```python
#Vols mes llargs
#reset i eliminacio de l'index en un nou dataframe
fitxer_ordenat=fitxer_selec.reset_index(drop=True)
#seleccionem les columnes origem destí i distancia
fitxer_ordenat=pd.DataFrame(fitxer_ordenat[['Origin','Dest','Distance']])
#agrupem els registres per origen, destí i distancia i ens quedem amb un dels registres del grup: el primer
fitxer_ordenat=fitxer_ordenat.groupby(['Origin','Dest','Distance']).first()
#ordenem el fitxer per distancia en ordre decreixent
fitxer_ordenat=fitxer_ordenat.sort_values('Distance', ascending=False)
#mostrem els 10 vols mes llargs
fitxer_ordenat.head(10)
```

```python
#Vols mes endarrerits
df_vols_mes_endarrerits=fitxer_selec.sort_values(by='ArrDelay',ascending=False)
df_vols_mes_endarrerits.head(10)
```

```python
#Rutes que acumulem mes retards
df_rutes_retard=fitxer_selec[['Origin','Dest','ArrDelay']]
df_rutes_retard=df_rutes_retard.reset_index(drop=True)
#fitxer_selec.groupby('UniqueCarrier')['Tard'].sum()
df_rutes_retard
df_rutes_agrupades=pd.DataFrame(df_rutes_retard.groupby(['Origin','Dest'])['ArrDelay'].sum())
df_rutes_agrupades=df_rutes_agrupades.sort_values(by='ArrDelay',ascending=False)
#presentem les 10 rutes amb mes retard acumulat
df_rutes_agrupades.head(10)
```
